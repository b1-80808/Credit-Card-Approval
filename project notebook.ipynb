## Oversampling

```python
df.drop(['NAME_HOUSING_TYPE','STATUS'], axis=1, inplace= True)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
X = df.drop(['TARGET'], axis=1)
Y = df['TARGET']

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, random_state=12346)
# Splitting the data into training and testing sets

# Assuming df is your DataFrame and 'target' is your target variable

# Now, you can perform oversampling on the training data
from imblearn.over_sampling import SMOTE

# Instantiate SMOTE
smote = SMOTE(random_state=42)

# Perform SMOTE oversampling on the training data only
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
def build_logistic_regression():
    from sklearn.linear_model import LogisticRegressionCV
    
    # create a model
    model = LogisticRegressionCV()

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
def build_knn_model():
    from sklearn.neighbors import KNeighborsClassifier

    # create a model
    model = KNeighborsClassifier(n_neighbors=7)

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
def build_decision_tree():
    # Create a decision tree classifier
    from sklearn.tree import DecisionTreeClassifier
    decision_tree = DecisionTreeClassifier()
    
    # Train the model
    decision_tree.fit(x_train, y_train)
    
    return decision_tree
```

```python
def build_random_forest():
    # Create a random forest classifier
    from sklearn.ensemble import RandomForestClassifier
    random_forest = RandomForestClassifier()
    
    # Train the model
    random_forest.fit(x_train, y_train)
    
    return random_forest
```

```python
def build_naive_bayes_model():
    from sklearn.naive_bayes import GaussianNB
    
    # create a model
    model = GaussianNB()

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
model_lg = build_logistic_regression()
```

```python
model_knn = build_knn_model()
```

```python
model_decision_tree = build_decision_tree()
```

```python
model_random_forest = build_random_forest()
```

```python
model_nb = build_naive_bayes_model()
```

```python
def evaluate_model(model):
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

    # define y_true and y_pred
    y_true = y_test
    y_pred = model.predict(x_test)    

    # get confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print(cm)

    # print accuracy
    print(f"accuracy = {accuracy_score(y_true, y_pred) * 100:.2f}%")
    print(f"precision = {precision_score(y_true, y_pred) * 100:.2f}%")
    print(f"recall = {recall_score(y_true, y_pred) * 100:.2f}%")
    print(f"F1 score = {f1_score(y_true, y_pred) * 100:.2f}%")
```

```python
evaluate_model(model_lg)
```

```python
evaluate_model(model_knn)
```

```python
evaluate_model(model_decision_tree)
```

```python
evaluate_model(model_random_forest)
```

```python
evaluate_model(model_nb)
```
